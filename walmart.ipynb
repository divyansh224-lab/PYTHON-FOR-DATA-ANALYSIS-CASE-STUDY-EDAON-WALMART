Q.2 WHAT IS THE AVERAGE UNIT PRICE OF ALL PRODUCT SOLD ?

```python
average_price = df.groupby('product_name')['unit_price'].mean()
average_price
```

```python
average_price = df.groupby('product_name')['unit_price'].mean()
plt.figure(figsize=(10, 6))
bars =average_price.plot(kind='bar',color = 'yellow')
plt.title('Average price per product')
plt.xlabel('Product Name')
plt.ylabel('Average Unit Price')
plt.xticks(rotation=45, ha ='center')
for index, value in enumerate(average_price):
  plt.text(index, value,f'{value:.2f}', ha='center', va='bottom',fontsize=10)
  plt.tight_layout()
plt.show()
```

Q.6 WHICH PRODUCT CATEGORIES SEE THE LARGEST INCRESE IN DEMAND DURING HOLIDAYS ?

```python
holiday_sales = df[df['holiday_indicator']== True]

holiday_demand = holiday_sales.groupby('product_name')['quantity_sold'].sum()

print(holiday_demand)
```

```python
from matplotlib.typing import LineStyleType
plt.figure(figsize=(12,6))
ax= holiday_demand.plot(kind='bar',color='skyblue',edgecolor='black')
for index, value in enumerate(holiday_demand):
  plt.text(index,value,f'{value}',ha='center',va='bottom',fontsize=14)

plt.title('total quantity sold during holidays byproduct', fontsize=14)
plt.xlabel('product name',fontsize=12)
plt.ylabel('total quantity sold',fontsize=12)
plt.xticks(rotation=45,ha='right',fontsize=10)
plt.grid(axis='y',linestyle='--',alpha=0.8)
plt.tight_layout()
plt.show()
```

Q.9 CALCULte the monthly sales and find out the trends

```python
# Convert transaction_date to datetime objects
df['transaction_date'] = pd.to_datetime(df['transaction_date'])



```

```python
# Extract Month Name and Month Number
df['Month'] = df['transaction_date'].dt.month_name()
df['Month_Number'] = df['transaction_date'].dt.month
```

```python
# Check unique months
df['Month'].unique()
```

```python


# Step 2: Group by Month and calculate total revenue
monthly_revenue = df.groupby('Month')['revenue'].sum().reset_index()

# Step 3: Plot the trend line
plt.figure(figsize=(10, 6))
plt.plot(monthly_revenue['Month'], monthly_revenue['revenue'], marker='o', linestyle='-', color='b')
plt.title('Monthly revenue Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.grid()
plt.show
```

Q.10 WHICH PRODUCT HAS THE BEST PERFORMANCE IN TERMS OF VOLUME SOLD

```python
best_product = df.groupby("product_name")["quantity_sold"].sum()
best_product
```

```python
# Plot the bar graph
plt.figure(figsize=(10, 6))
best_product.plot(kind='bar', color='orange', edgecolor='black')

# Customize the plot
plt.title('Best Selling Products', fontsize=14)
plt.xlabel('Product Name', fontsize=12)
plt.ylabel('Quantity Sold', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Add data labels
for index, value in enumerate(best_product):
    plt.text(
        index, value + 0.01 * max(best_product),  # Slightly above the bar
        f'{value:.0f}',  # Format the value
        ha='center', va='bottom', fontsize=10, color='black'
    )
```
